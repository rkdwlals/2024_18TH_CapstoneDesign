```python
import cv2
import math
import numpy as np
from skimage.draw import line
import os
import size_equalize
IMAGESIZE = 1000

os.chdir("C:/Users/user/Desktop/scoreDraw")
# print(os.getcwd())
```

```python

exp_img_circle = cv2.imread("./CircleExample.jpg")
# draw_img_circle = cv2.imread("./SquareDraw.jpg")
# draw_img_circle = cv2.imread("./CircleDraw.jpg")
draw_img_circle = cv2.imread("./CircleDrawNew.jpg")
# draw_img_circle = cv2.imread("./CircleDrawDisconnect.jpg")
exp_img_square = cv2.imread("./SquareExample.jpg")
draw_img_square = cv2.imread("./SquareDraw.jpg")
# draw_img_square = cv2.imread("./SquareDrawDisconnect.jpg")
# draw_img_square = cv2.imread("./SquareDrawDisconnect4.jpg")
# draw_img_square = cv2.imread("./SquareDrawEdit2.jpg")

draw_img_circle = cv2.resize(draw_img_circle, (IMAGESIZE, IMAGESIZE))
exp_img_circle = cv2.resize(exp_img_circle, (IMAGESIZE, IMAGESIZE))
draw_img_square = cv2.resize(draw_img_square, (IMAGESIZE, IMAGESIZE))
exp_img_square = cv2.resize(exp_img_square, (IMAGESIZE, IMAGESIZE))

exp_org_circle = exp_img_circle.copy()
draw_org_circle = draw_img_circle.copy()
exp_org_square = exp_img_square.copy()
draw_org_square = draw_img_square.copy()


# cv2.imshow("ec",exp_img_circle)
# cv2.imshow("dc",draw_img_circle)
# cv2.imshow("es",exp_img_square)
# cv2.imshow("ds",draw_img_square)

# cv2.waitKey()
# cv2.destroyAllWindows()
```

```python
# 잘라낼 좌표점
# 각 좌표가 비교 가능하게 초기화
# 좌표값 반환 함수 (4 꼭짓점 좌표 리스트 반환 )
left_up_exp_circle, left_down_exp_circle, right_up_exp_circle, right_down_exp_circle = size_equalize.get_extreme_points(exp_img_circle)
left_up_draw_circle, left_down_draw_circle, right_up_draw_circle, right_down_draw_circle = size_equalize.get_extreme_points(draw_img_circle)
left_up_exp_square, left_down_exp_square, right_up_exp_square, right_down_exp_square = size_equalize.get_extreme_points(exp_img_square)
left_up_draw_square, left_down_draw_square, right_up_draw_square, right_down_draw_square = size_equalize.get_extreme_points(draw_img_square)
```

```python

# perspective Transform ( 예시 이미지 '원' )
orignal_pnt_exp_circle = np.float32([left_up_exp_circle, left_down_exp_circle, right_up_exp_circle, right_down_exp_circle])
move_pnt = np.float32([[0, 0], [0, 1000], [1000, 0], [1000, 1000]])

perspective_exp_circle = cv2.getPerspectiveTransform(orignal_pnt_exp_circle, move_pnt)
dst_exp_circle = cv2.warpPerspective(exp_org_circle.copy(), perspective_exp_circle, (1000, 1000))

# perspective Transform ( 아동 이미지 '원' )
orignal_pnt_draw_circle = np.float32([left_up_draw_circle, left_down_draw_circle, right_up_draw_circle, right_down_draw_circle])

perspective_draw_circle = cv2.getPerspectiveTransform(orignal_pnt_draw_circle, move_pnt)
dst_draw_circle = cv2.warpPerspective(draw_org_circle.copy(), perspective_draw_circle, (1000, 1000))

# perspective Transform ( 예시 이미지 '사각형' )
orignal_pnt_exp_square = np.float32([left_up_exp_square, left_down_exp_square, right_up_exp_square, right_down_exp_square])

perspective_exp_square = cv2.getPerspectiveTransform(orignal_pnt_exp_square, move_pnt)
dst_exp_square = cv2.warpPerspective(exp_org_square.copy(), perspective_exp_square, (1000, 1000))

# perspective Transform ( 아동 이미지 '사각형' )
orignal_pnt_draw_square = np.float32([left_up_draw_square, left_down_draw_square, right_up_draw_square, right_down_draw_square])

perspective_draw_square = cv2.getPerspectiveTransform(orignal_pnt_draw_square, move_pnt)
dst_draw_square = cv2.warpPerspective(draw_org_square.copy(), perspective_draw_square, (1000, 1000))

# cv2.imwrite("./result_image/draw_circle_cutresult.jpg", dst_draw_circle)
# cv2.imshow("exp circle", dst_exp_circle)
# cv2.imshow("draw circle", dst_draw_circle)
# cv2.imshow("exp square", dst_exp_square)
# cv2.imshow("draw square", dst_draw_square)
# cv2.waitKey()
# cv2.destroyAllWindows()
```

```python
import circle_score
circle_score_result = circle_score.circle_score_result(dst_draw_circle, dst_exp_circle)

print(circle_score_result)
```
